```python
import pandas as pd
# Set the display options
pd.set_option('display.max_rows', None)  # No limit on the number of rows displayed
pd.set_option('display.max_columns', None)  # No limit on the number of columns displayed
pd.set_option('display.width', None)  # Auto-detect the display width for wrapping
pd.set_option('display.max_colwidth', None)  # No truncation on the column width

# Load JSON data from a URL
file_path = f'D:\\Project\\Jupyter\\Stock\\Main\\Result\\checkall\\2025-03-05.xlsx'
df_file = pd.read_excel(file_path)
df_file=df_file[df_file['now_price']>0]
```

```python
df_file['now_price']=df_file['now_price'].replace(",", "").astype(float)

df_file['每股淨值(元)'] =  df_file['每股淨值(元)'].astype(str).str.strip()   # 去掉前後空白
df_file['每股淨值(元)'] = df_file['每股淨值(元)'].str.replace(",", "", regex=True).astype(float)
df_file['淨值倍率']=(df_file['now_price']/df_file['每股淨值(元)']).astype(float).apply(lambda x: f"{x:.2f}")
```

```python
file_path = f'D:\\Project\\Jupyter\\Stock\\Main\\Result\check1.xlsx'
df = pd.read_excel(file_path)

# 檢查檔案中的欄位名稱
print(df.columns)

# 設定篩選條件：
# 1. 成交量大於 5000
# 2. 波動幅度（漲跌幅 / 現價）大於 3%
# 3. 本益比大於 0，確保公司有獲利

# 計算波動幅度（假設 change_quote 是漲跌幅、now_price 是現價）
df['波動幅度'] = abs(df['change_quote'] / df['now_price'] * 100)

# 篩選符合條件的股票
day_trade_stocks = df[
    (df['成交量'] > 5000) &
    (df['波動幅度'] > 3) &
    (df['本益比'] > 0)
]

# 挑選重點欄位並按成交量降冪排序
top_day_trade_stocks = day_trade_stocks.sort_values(by='成交量', ascending=False)[
    ['stock_number', 'Type0', 'Type1', 'now_price', '成交量', '波動幅度', '本益比', '總市值']
]

# 顯示篩選後前 10 名
print(top_day_trade_stocks.head(30))
```

```python


# 1. 著重觀察哪些產業趨勢（統計各產業出現在篩選結果中的次數）
industry_trend = day_trade_stocks['Type1'].value_counts()
print("產業趨勢觀察：")
print(industry_trend)

# 2. 篩選低檔股票（假設以本益比與股價低於中位數作為低檔標準）
low_price_stocks = df[
    (df['now_price'] < df['now_price'].median()) &
    (df['本益比'] < df['本益比'].median())
]

# 挑選重點欄位並按本益比升冪排序
low_price_stocks_sorted = low_price_stocks.sort_values(by='本益比')[
    ['stock_number', 'Type0', 'Type1', 'now_price', '成交量', '波動幅度', '本益比', '總市值']
]

# 顯示低檔觀察標的前 10 名
print("低檔股票觀察標的：")
print(low_price_stocks_sorted.head(10))
```
